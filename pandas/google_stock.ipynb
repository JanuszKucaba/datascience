```python
import pandas as pd
import pandas_datareader.data as web
import plotly.express as px

pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df = web.DataReader(name='GOOGL', data_source='stooq')
df
```

```python
df.describe()
```

```python
df = df.reset_index()
df.head()
```

```python
df.tail()
```

```python
df_2019 = df.query("Date > '2019-01-01'")
df_2019
```

```python
px.line(df_2019, x='Date', y='Close', title='Stock quotes of the google company (GOOGL)',
        width=900, height=500, log_y=True, color_discrete_sequence=['#03fcc2'],
        template='plotly_dark')
```

```python
px.area(df_2019, x='Date', y='Close', title='Stock quotes of the google company (GOOGL)',
        width=900, height=500, log_y=True, color_discrete_sequence=['#c50ff7'],
        template='plotly_dark')
```
